# Análise Preditiva

A análise preditiva é o processo de usar dados para prever resultados futuros.

O processo usa análise de dados, machine learning, inteligência artificial e modelos estatísticos para encontrar padrões que possam prever comportamentos futuros.

As organizações podem usar dados históricos e atuais para prever tendências e comportamentos com segundos, dias ou anos de antecedência, com muita precisão.

Elas podem ser classificadas de duas formas:

* Os **modelos de classificação** tentam colocar objetos de dados (como clientes ou possíveis resultados) em uma categoria ou outra. Por exemplo, se um varejista tiver muitos dados sobre diferentes tipos de cliente, ele poderá tentar prever quais tipos de cliente serão receptivos a e-mails de marketing.
* Os **modelos de regressão** tentam prever dados contínuos, por exemplo, a quantidade de receita que um cliente vai gerar durante seu relacionamento com uma empresa.

A principal diferença entre classificação e regressão é o tipo de saída que cada uma produz. A classificação produz uma saída discreta, enquanto a regressão produz uma saída contínua.

Com relação às técnicas utilizadas, podemos classificar da seguinte forma:

* **Análise de regressão**: A regressão é uma técnica de análise estatística que estima as relações entre variáveis. Ela é útil para determinar padrões em grandes conjuntos de dados e, assim, determinar a correlação entre as entradas. É melhor usá-la em dados contínuos que sigam uma distribuição conhecida. A regressão é frequentemente usada para determinar como uma ou mais variáveis independentes afetam outra, por exemplo, como um aumento de preço afetará a venda de um produto.

* **Árvores de decisão**: Árvores de decisão são modelos de classificação que colocam dados em diferentes categorias com base em variáveis distintas. O método é o mais eficaz para entender as decisões de uma pessoa. O modelo lembra uma árvore, com cada ramificação representando uma possível opção e a folha associada, o resultado da decisão. Normalmente, as árvores de decisão são fáceis de entender e funcionam quando um conjunto de dados tem diversas variáveis faltando.

* **Redes neurais**: Redes neurais são métodos de machine learning úteis em análises preditivas que modelam relações muito complexas. Basicamente, elas são potentes mecanismos de reconhecimento de padrões. As redes neurais são mais adequadas para determinar relações não lineares em conjuntos de dados, especialmente quando não existe nenhuma fórmula matemática conhecida para analisá-los. As redes neurais podem ser usadas para validar os resultados das árvores de decisão e modelos de regressão.

## Regressão Linear

Dentre as técnicas de regressão, a mais simples é a Linear.

Trata-se de uma técnica estatística usada para encontrar a relação entre variáveis. Em um contexto de ML, a regressão linear encontra a relação entre **atributos** (também chamado de **features**) e um **rótulo** (também chamado de **target**).

A fórmula utilizada para a previsão é a seguinte:

$y' = b + w_1 x_1$

sendo que:
* $y'$ é o valor que estamos prevendo (target);
* $b$ é chamado de $bias$ ou viés, trata-se do ponto em que a linha toca o eixo y;
* $w_1$ é o peso ou ponderação, determina a inclinação da linha;
* $x_1$ são os valores da feature.

É possível utilizar mais de uma feature para o cálculo, sendo que a fórmula utilizada é a seguinte:

$y' = b + w_1 x_1 + w_2 x_2 + w_3 x_3 + ... + w_n x_n$

A **perda** é uma métrica numérica que descreve o quão erradas são as previsões de um modelo. A perda mede a distância entre as previsões do modelo e os rótulos reais. O objetivo de treinar um modelo é minimizar a perda, reduzindo-a ao menor valor possível.

Existem várias fórmulas para calcular a perda. Uma das mais usadas é o **MSE** (traduzindo para **Erro quadrático médio**). Segue a fórmula:

$\frac{1}{N} ∑ (valor\ atual - valor\ previsto)^2$

Para calcular os melhores valores de Viés e Peso, utilizamos um algoritmo chamado de **Gradiente Descendente**.

Funciona da seguinte forma: O modelo começa o treinamento com pesos e vieses aleatórios próximos de zero e repete as seguintes etapas:
1. Calcula a perda com o peso e o viés atuais;
2. Determina a direção para mover os pesos e o viés que reduzem a perda;
3. Move os valores de peso e viés um pouco na direção que reduz a perda;
4. Volta à etapa 1 e repita o processo até que o modelo não consiga mais reduzir a perda.

Ao treinar um modelo, você geralmente analisa uma **curva de perda** para determinar se o modelo convergiu. A curva de perda mostra como a perda muda à medida que o modelo é treinado.

Para melhorar a velocidade e a precisão do modelo, podemos ajustar os **hiperparâmetros** do treinamento. Os mais conhecidos são:
* A **taxa de aprendizado** determina a magnitude das mudanças a serem feitas nos pesos e no viés durante cada etapa do processo de descida do gradiente. Se muito baixa, o modelo pode levar muito tempo para convergir. No entanto, se muito alta, o modelo nunca vai convergir, mas vai variar em torno dos pesos e do viés que minimizam a perda.
* O **tamanho do lote** é um hiperparâmetro que se refere ao número de exemplos que o modelo processa antes de atualizar os pesos e o viés.
* Durante o treinamento, uma **época** significa que o modelo processou uma vez cada exemplo no conjunto de treinamento. O treinamento geralmente requer muitas épocas. Ou seja, o sistema precisa processar cada exemplo no conjunto de treinamento várias vezes. Em geral, mais épocas produzem um modelo melhor, mas também levam mais tempo para treinar.

## Exemplo Prático

Vamos a um exemplo de código para calcular a regressão linear. Segue abaixo os dados utilizados:

```python
import matplotlib.pyplot as plt
from scipy import stats

ageCar = [5,7,8,7,2,17,2,9,4,11,12,9,6]
maxSpeedCar = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(ageCar, maxSpeedCar)
plt.xlabel("Idade do Carro")
plt.ylabel("Velocidade Máxima")
plt.show()
```

Segue código para calcular a regressão linear e os detalhes do treinamento.

```python
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

X = np.reshape(ageCar, (-1, 1))
y = np.reshape(maxSpeedCar, (-1, 1))

# Construir regressão linear
reg = LinearRegression()
reg.fit(X, y)

print("f(x) = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))

# Plotar gráfico
f_previsaoes = reg.predict(X)
plt.figure(figsize = (16,8))
plt.scatter(
    ageCar,
    maxSpeedCar,
    c='red')

plt.plot(
    ageCar,
    f_previsaoes,
    c='blue',
    linewidth=3,
    linestyle=':'
)

plt.xlabel("Idade do Carro")
plt.ylabel("Velocidade Máxima")
plt.show()

# Apresentar resumo
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

### Regressão Linear Múltipla

Segue um exemplo com regressão linear com múltiplas features:

```python
from sklearn.linear_model import LinearRegression
import numpy as np

hp = np.array([75,80,65,90,101,76,66,77,80,76,120,120,73])

Xs = np.array((ageCar, hp))
Xs = np.reshape(Xs, (-1, 2))

# Construção do Modelo
modelo = LinearRegression()
modelo.fit(Xs, y)
print("f(x) = {:.5} + {:.5}*X1 + {:.5}*X2".format(modelo.intercept_[0], modelo.coef_[0][0], modelo.coef_[0][1]))

X2 = sm.add_constant(Xs)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
