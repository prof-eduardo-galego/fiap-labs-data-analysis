Faz a análise de dados da Folha de Pagamento dos servidores públicos do Estado de São Paulo, com objetivo de procurar padrões sobre os dados e possíveis desvios ou irregularidades.

O dataset está disponível em: https://www.transparencia.sp.gov.br/PortalTransparencia-Report/txt/RemuneracaoAtivos.csv

Dica: Faça o download do arquivo e importe apenas os primeiros 10 registros para testar os comandos. Depois que tudo estiver correto, importe o arquivo completo.

```python
import pandas as pd                         # Biblioteca para trabalhar com datasets
import numpy as np                          # Biblioteca de funções matemáticas
import matplotlib.pyplot as plt             # Biblioteca para criação de gráficos (simples)
import seaborn as sns                       # Biblioteca para criação de gráficos (avançado)

%matplotlib inline
sns.set(color_codes=True)

df = pd.read_csv("https://www.transparencia.sp.gov.br/PortalTransparencia-Report/txt/RemuneracaoAtivos.csv", encoding='latin-1', sep=';')
display(df.head(5))
print(df.shape)
```

Segue um trecho de código para auxiliar na transformação dos dados numéricos:

```python
columns_to_convert = [
    'FÉRIAS E 13º SALÁRIO',
    'PAGAMENTOS EVENTUAIS',
    'LICENÇA PRÊMIO INDENIZADA',
    'ABONO PERMANÊNCIA & OUTRAS INDENIZAÇÕES',
    'REDUTOR SALARIAL',
    'TOTAL LIQUÍDO (R$)'
]

for col in columns_to_convert:
    df[col] = df[col].str.replace(',', '.', regex=False).astype(float)

# Display the data types to confirm the changes
display(df.dtypes)
```

Segue um exemplo de código que efetua a soma do total líquido agrupado por órgão:

```python
# Calculate the sum of 'TOTAL LIQUÍDO (R$)' grouped by 'ORGÃO'
total_liquido_por_orgao = df.groupby('ORGÃO')['TOTAL LIQUÍDO (R$)'].sum()

# Display the result
display(total_liquido_por_orgao)
```
